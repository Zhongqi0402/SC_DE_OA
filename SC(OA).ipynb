```python
import pandas as pd
```

```python
data = 'Airline Code;DelayTimes;FlightCodes;To_From\nAir Canada (!);[21, 40];20015.0;WAterLoo_NEWYork\n<Air France> (12);[];;Montreal_TORONTO\n(Porter Airways. );[60, 22, 87];20035.0;CALgary_Ottawa\n12. Air France;[78, 66];;Ottawa_VANcouvER\n""".\\.Lufthansa.\\.""";[12, 33];20055.0;london_MONTreal\n'
```

### Converting stringified data to a pandas dataframe

```python
rows = data.split('\n')
data = []
for each in rows:
    data.append(each.split(';'))
df = pd.DataFrame(data[1:len(data)-1], columns=data[0])
df
```

### Requirement 1

```python
df['FlightCodes'] = pd.to_numeric(df['FlightCodes'])
flightCode = list(df['FlightCodes'])
for i, each in enumerate(flightCode):
    if pd.isna(each): flightCode[i] = int(flightCode[i-1])+10
    else: flightCode[i] = int(each)
df['FlightCodes'] = flightCode
df
```

### Requirement 2

```python
df[['To', 'From']] = df["To_From"].str.upper().str.split("_", expand=True)
df = df.drop('To_From', axis=1)
df
```

### Requirement 3

```python
airlineCode = list(df['Airline Code'])
for i, each in enumerate(airlineCode):
    s = ''.join(ch for ch in each if ch.isalpha() or ch==' ')
    airlineCode[i] = s
stripped = list(map(str.strip, airlineCode))
df['Airline Code'] = stripped
df
```
